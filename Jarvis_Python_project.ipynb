```python
import speech_recognition as sr
import webbrowser
import pyttsx3
import requests
import openai
import re


# Initialize the recognizer and text-to-speech engine
recognizer = sr.Recognizer()
engine = pyttsx3.init()

openai.api_key = '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'
weather_api_key = '0d3e1f60b71cdda089d61732d33f15e3'

def speak(text):
    """Function to convert text to speech."""
    engine.say(text)
    engine.runAndWait()


def get_openai_response(prompt):
    """Function to get a response from OpenAI's GPT model."""
    response = openai.Completion.create(
        engine="gpt-3.5-turbo",  
        prompt=prompt,
        max_tokens=100
    )
    return response.choices[0].text.strip()

def get_weather(weather_api_key):
    speak("Please say the name of the city you want the weather for.")
    print("Listening for city name...")
    
    with sr.Microphone() as source:
        audio = recognizer.listen(source)
    try:
        city_name = recognizer.recognize_google(audio).lower()
        print(f"City name recognized: {city_name}")
        
        base_url = f"http://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={api_key}&units=metric"
        response = requests.get(base_url)
        data = response.json()

        if "main" in data:
            main = data["main"]
            weather_desc = data["weather"][0]["description"]
            temperature = main["temp"]
            weather_report = f"The weather in {city_name} is {weather_desc} with a temperature of {temperature}Â°C."
            return weather_report
       
        else:
            error_message = f"Error: Unable to retrieve weather data for {city_name}."
            if "message" in data:
                error_message += f" API Error Message: {data['message']}"
            return error_message
            
    except sr.UnknownValueError:
        return "Sorry, I couldn't understand the city name."
    except sr.RequestError as e:
        return f"Could not request results from Google Speech Recognition service; {e}"
        
def calculate_expression(expression):
    """Function to evaluate mathematical expressions."""
    try:
        # Remove any non-numeric and non-operator characters
        expression = re.sub(r'[^0-9+\-*/().]', '', expression)
        result = eval(expression)
        return f"The result of {expression} is {result}."
    except Exception as e:
        return f"Sorry, I couldn't calculate that. Error: {e}"
        
def processCommand(command):
    """Process voice commands and perform actions."""
    command = command.lower()

    print(f"Command received: {command}")

    if "open google" in command:
        webbrowser.open("https://google.com")
    elif "open facebook" in command:
        webbrowser.open("https://facebook.com")
    elif "open youtube" in command:
        webbrowser.open("https://youtube.com")
    elif "open linkedin" in command:
        webbrowser.open("https://linkedin.com")
    elif "open spotify" in command:
        webbrowser.open("https://open.spotify.com/track/74OuU0iGq2Hbsgnetf5xhy?si=7017e9c56d03409c")
    elif "open news" in command:
        webbrowser.open("https://news.google.com/search?q=crab&hl=en-US&gl=US&ceid=US%3Aen")
        speak("Opening Google News")
    elif "open weather report" in command:
        weather_report = get_weather(weather_api_key) 
        speak(weather_report)
    elif any(op in command for op in ["+", "-", "*", "/"]):
        result = calculate_expression(command)
        speak(result)
    else:
        response = get_openai_response(command)
        speak(response)
        print(response)

    
if __name__ == "__main__":
    speak("Initializing Jarvis...")
    while True:
        try:
            # Listen for the wake word "Jarvis"
            print("Listening for wake word 'Jarvis'...")
            with sr.Microphone() as source:
                audio = recognizer.listen(source, timeout=4)
            word = recognizer.recognize_google(audio).lower()

            if word == "jarvis":
                speak("Ya")
                print("Jarvis is listening...")
                with sr.Microphone() as source:
                    audio = recognizer.listen(source)
                    command = recognizer.recognize_google(audio)
                    processCommand(command)                  

        except Exception as e:
            print(f"Error: {e}")
```
